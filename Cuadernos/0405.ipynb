# Ejercicio final del tema 4 - Red social
En temas anteriores, el usuario no tenía ninguna capacidad de decisión. Solo iba respondiendo a lo que el programa le pedía. Ahora vamos a dejarle tomar pequeñas decisiones, implantando un sencillo menú.

Modifica el programa anterior de forma que:
- Al principio del todo, le pregunte al usuario si quiere crear un perfil nuevo o no. El usuario debe poder contestar a esta pregunta con variantes de `sí` y de `no`, como `s`, `SÍ`, `sip`, `YES`, etc. Basta con que pongas dos para cada respuesta posible, pero puedes dejar volar tu imaginación.
    - Si quiere crearlo, usaremos el código que ya tenemos escrito en el ejercicio del tema 3 para pedirle sus datos e imprimirlos por pantalla, diciendo `Tu perfil ha sido creado`. Pero nada más; ya no le pediremos un mensaje de estado después.
    - Si el usuario no quiere crear un perfil nuevo, mostraremos entonces un menú con 3 opciones, dándole a entender que si escribe `1` elige la primera opción, si escribe `2` la segunda, etcétera. Las opciones serán:
        - Crear un perfil nuevo: en este caso, reutilizaremos el código con el que le pedíamos sus datos y los imprimíamos por pantalla, diciendo `Tu perfil ha sido creado`.
        - Modificar sus datos: para este caso, también reutilizaremos el código con el que le pedíamos sus datos y los imprimíamos por pantalla, pero esta vez diremos `Tu perfil ha sido actualizado`.
        - Publicar un mensaje: en este caso, usaremos el código que habíamos escrito para preguntarle qué quiere publicar e imprimirlo por pantalla. Como en este caso no nos ha dado ningún nombre, pondremos algo como `El usuario dice:` antes de imprimir su mensaje.
