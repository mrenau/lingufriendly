- Existan las siguientes funciones:

- Esté separado en 3 partes: datos (que contendrá el diccionario de pronvincias), funciones y programa principal, en ese orden.
- Esté bien documentado, es decir, haya comentarios separando las partes del programa y lo que hace cada función, al menos.
